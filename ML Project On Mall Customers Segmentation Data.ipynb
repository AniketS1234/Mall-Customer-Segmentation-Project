## Segmentating Data Based On Different Category Based On Spending Score

```python
conditions = [
    (df['Spending Score (1-100)'] < 20),
    (df['Spending Score (1-100)'] >= 20) & (df['Spending Score (1-100)'] < 35),
    (df['Spending Score (1-100)'] >= 35) & (df['Spending Score (1-100)'] < 50),
    (df['Spending Score (1-100)'] >= 50) & (df['Spending Score (1-100)'] < 65),
    (df['Spending Score (1-100)'] >= 65) & (df['Spending Score (1-100)'] < 80), 
    (df['Spending Score (1-100)'] >= 80),
]

values = ['Miser','General','Careful','Target','Spendthrift','Centroid']

df['Category_Group'] = np.select(conditions,values)
print(df)
```

## KNN Algorithm

```python
x = as1[['Annual Income (k$)', 'Spending Score (1-100)']]
y = as1['Category_Group']
```

```python
from  sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
y_test
```

```python
y_train
```

```python
x_train
```

```python
x_test
```

```python
knn = KNeighborsClassifier()
```

```python
knn.fit(x_train,y_train)
```

```python
knn.score(x_test,y_test)
```

**Above Model Score Is Achieved For Different Category Segment Groups Based On Annual Income And Spending Score Using KNN Algorithm.**

```python
k_range = range(1,31)
score = []
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(x_train,y_train)
    score.append(knn.score(x_test,y_test))
plt.xlabel("k")
plt.ylabel("Accuracy")
plt.legend()
plt.scatter(k_range,score)
plt.show()
```

## Decision Tree Algorithm

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc1 = RandomForestClassifier(random_state=0)
```

```python
rfc1.fit(x_train,y_train)
```

```python
y_pred = rfc1.predict(x_test)
```

```python
y_pred
```

**Above Model Score Is Achieved For Different Category Segment Groups Based On Annual Income And Spending Score Using Decision Tree Algorithm.**

```python
k_range = range(1,31)
score = []
for k in k_range:
    rfc1 = RandomForestClassifier(random_state=0)
    rfc1.fit(x_train,y_train)
    score.append(rfc1.score(x_test,y_test))
plt.xlabel("rfc1")
plt.ylabel("Accuracy")
plt.legend()
plt.scatter(k_range,score)
plt.show()
```
